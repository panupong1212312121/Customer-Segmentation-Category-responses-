# What are the characteristics of each Category?

```python
# qualitative features 
feat_qual = df[['Gender', 'Married', 'Graduated', 'Profession','SpendingScore']]

# quantitative features 
feat_quan = df[['Age','WorkExperience', 'FamilySize']]
```

Consider qualitative features and Category cluster 

Objective : I want to know each group has different proportion value in each feature or not, So I decide to conduct criteria below :

1.Use proportions_ztest when assumption are met 

2.Use fisher_exact when assumption aren't met 

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import fisher_exact

qual_sig = []
qual_insig = []

for i in feat_qual.columns:
    X = df[[i,'Category']]
    X = X.dropna()
    X.reset_index(inplace=True)
    X.drop('index',axis=1,inplace=True)
    category = ['Category 1', 'Category 2', 'Category 3', 'Category 4','Category 5', 'Category 6', 'Category 7']
    feat = X[i].unique()
    for j in range(len(category)-1):
        for k in range(j+1,len(category)):
            cat1 = X[X['Category']==category[j]]
            cat2= X[X['Category']==category[k]]
            for l in range(len(feat)):
                # define the sample sizes and proportions for the two groups
                n1 = len(cat1)
                n2 = len(cat2)
                feat1 = cat1[cat1[i]==feat[l]]
                feat2 = cat2[cat2[i]==feat[l]]
                p1 = len(feat1)/n1
                p2 = len(feat1)/n1

                name = category[j]+' VS '+category[k]+' : '+feat[l]
                print(name)

                # check assumptions 
                if (n1*p1 >= 5 and n1*(1-p1) >= 5 and n2*p2 >= 5 and n2*(1-p2) >= 5):
                    # perform the two-sample z-test for the difference in proportions
                    print("Assumptions are met")
                    z_stat, p_val = proportions_ztest(count= [n1*p1, n2*p2], nobs=[n1, n2])
                    if p_val < 0.05:
                        print('There is significantly different with p-value = {}'.format(p_val))
                        qual_sig.append(name)
                    else:
                        print('There is insignificantly different with p-value = {}'.format(p_val))
                        qual_insig.append(name)
                else:
                    print("Assumptions are not met")

                    # calculate the observed number of successes and failures for each group
                    obs_successes = np.array([int(n1*p1), int(n2*p2)])
                    obs_failures = np.array([int(n1*(1-p1)), int(n2*(1-p2))])

                    # perform the fisher_exact for the difference in proportions
                    odds_ratio, p_val = fisher_exact([obs_successes, obs_failures])
                    if p_val < 0.05:
                        print('There is significantly different with p-value = {}'.format(p_val))
                        qual_sig.append(name)
                    else:
                        print('There is insignificantly different with p-value = {}'.format(p_val))
                        qual_insig.append(name)
                print('-'*40)
            print('-'*70)
        print('-'*100)

print('qual_sig : \n {}'.format(qual_sig))
print('qual_insig : \n {}'.format(qual_insig))
```

From above , each category is insignificantly different proportion in each qualitative feature 

So I summarize that each category isn't different proportion in qualitative feature

Consider quantitative features and Category cluster 

Objective : I want to know each group has different location in each feature or not, So I decide to conduct criteria below :

1.Use Wilcoxon rank sum test when data is non-normal

2.Use Z-test (n1 and n2 >= 30) or t-test (At least n1 or n2 <30) when data is normal

```python
feat_quan
```

```python
from scipy.stats import kstest , shapiro

quan_normal = []
quan_non_normal = []

for i in feat_quan.columns:
    X = df[[i,'Category']]
    X = X.dropna()
    X.reset_index(inplace=True)
    X.drop('index',axis=1,inplace=True)
    category = ['Category 1', 'Category 2', 'Category 3', 'Category 4','Category 5', 'Category 6', 'Category 7']
    for j in range(len(category)):
        data = X[X['Category']==category[j]]
    
        name = category[j] + ' : '+i
        print(name)

        if len(data) > 50:

            statistics , pvalue = kstest(data[i],'norm')
            print('len = {}'.format(len(data)))
            print('len > 50 : use kstest')
            print('p-value : {}\n' .format(pvalue))

            if pvalue < 0.05:
                print('Non-Normal')
                quan_non_normal.append(name)

            else:
                print('normal')
                quan_normal.append(name)

        else:
            statistics , pvalue = shapiro(data[i])
            print('len = {}'.format(len(data)))
            print('len <= 50 : use shapiro')
            print('p-value : {}\n' .format(pvalue))
            
            if pvalue < 0.05:
                print('Non-Normal')
                quan_non_normal.append(name)

            else:
                print('normal')
                quan_normal.append(name)

        print('-'*40)
            
print('quan_normal : \n {}'.format(quan_normal))
print('quan_non_normal : \n {}'.format(quan_non_normal))
```

From above , I found each category is non-normal -> Using wilcoxon rank sum test 

```python
# Check each group has different (greater or less) location or not?

from scipy.stats import ranksums

list_same = []
list_greater = []
list_less = []

for i in feat_quan.columns:
    X = df[[i,'Category']]
    X = X.dropna()
    X.reset_index(inplace=True)
    X.drop('index',axis=1,inplace=True)
    category = ['Category 1', 'Category 2', 'Category 3', 'Category 4','Category 5', 'Category 6', 'Category 7']
    for j in range(len(category)-1):
        for k in range(j+1,len(category)):
            cat1 = X[X['Category']==category[j]]
            cat2 = X[X['Category']==category[k]]

            statistics , pvalue = ranksums(cat1[i],cat2[i])

            name = category[j] + ' VS ' + category[k]+' : '+i
            print(name)

            if pvalue<0.05:
                statistics , pvalue = ranksums(cat1[i],cat2[i],alternative='greater')
                if pvalue<0.05:
                    print('Greater')
                    list_greater.append(name)
                else:
                    print('Less')
                    list_less.append(name)
            else:
                print('p-value : {}'.format(pvalue))
                print('Same')
                list_same.append(name)
            print('-'*40)

print('Same : \n{} \n'.format(list_same))
print('Greater : \n{} \n'.format(list_greater))
print('Less : \n{}'.format(list_less))
```

From above , I get result like those 

You can interpret like below :

E.g. Same :

Category 1 VS Category 2 : Age

Summary : age of Category 1 is same as Category 2

E.g. Greater :

Category 6 VS Category 7 : Age 

Summary : age of Category 6 greater than Category 7

E.g. Less :

Category 2 VS Category 6 : Age 

Summary : age of Category 2 less than Category 6

# Build model to predict unknown category

```python
df = pd.read_csv('Customer Segmentation.csv')
df.dropna(inplace=True)
df.drop('CustomerID',axis=1,inplace=True)
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
X = df.drop(['Category'],axis=1) #Features
y = df[['Category']] #Category
```

Before training model , you must convert categorical features to numeric by using some technique 

I choose get dummy technique to perform one-hot encoding

```python
X = pd.get_dummies(X)
X
```

```python
# create a dictionary to map old values to new values
cat_dict = {'Category 1': 1, 'Category 2': 2, 'Category 3': 3, 'Category 4': 4, 'Category 5': 5, 'Category 6': 6, 'Category 7': 7}

# use replace() method to replace old values with new values
y['Category'] = y['Category'].replace(cat_dict)
y
```

```python
for i in y['Category'].unique():
    print('Category {}'.format(i))
    print('count = {} '.format(len(y[y['Category']==i])))
    print('-'*40)
```

From above , I found my data is imbalance , so I decide to use techniques which can handle imbalance problem like below:

1.SMOTE

2.Class_weight 

```python
from imblearn.over_sampling import SMOTE

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import cross_val_score

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

smote = SMOTE(random_state=0)

X_scaled = scaler.fit_transform(X)

# Resample the dataset
X_resampled, y_resampled = smote.fit_resample(X_scaled, y.values.ravel())

# Candidated algorithm models
models = []
models.append(('LR', LogisticRegression(class_weight='balanced', 
                                        multi_class='multinomial',
                                        max_iter=1000)))
models.append(('kNN', KNeighborsClassifier()))
models.append(('SVC', SVC(class_weight='balanced')))
models.append(('RFC', RandomForestClassifier(class_weight='balanced', 
                                             random_state=0)))
models.append(('DT', DecisionTreeClassifier(class_weight='balanced', 
                                            random_state=0)))
models.append(('GNB', GaussianNB()))

names = []
results = []

for name, model in models:
    scores = cross_val_score(model, X_resampled, y_resampled, cv=10, scoring='accuracy')
    names.append(name)
    results.append(scores)
    print('{:4} -> Accuracy : {:.3f} % (Std : {:.3f} %)'.format(name, scores.mean()*100, scores.std()*100))
```

From above , I choose RFC model for running in checking overfitting process because it has maximum accuracy

```python
from sklearn.model_selection import learning_curve

rfc =  RandomForestClassifier(class_weight='balanced', 
                                random_state=0)

train_sizes, train_scores, test_scores = learning_curve(rfc,
                                                        X_resampled, y_resampled, cv=10, train_sizes=np.linspace(.1, 1.0, 5),scoring='accuracy')

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
plt.title('RFC')
plt.ylabel('Accuracy')
plt.xlabel('Sizes')
plt.legend(loc="best")
plt.show()
```

As you can see from graph above , there is no large gap between these two lines 

I summarize that my model isn't overfitting 

The next step : Tuning model for improving accuracy

```python
from sklearn.model_selection import GridSearchCV

# Define the range of hyperparameters to optimize
param_grid = {'criterion':['gini', 'entropy', 'log_loss'],
                'max_features':['sqrt', 'log2', None],
                'class_weight':['balanced', 'balanced_subsample',None]
                }

# Perform the grid search
grid_search = GridSearchCV(RandomForestClassifier(random_state=0), 
                            param_grid, 
                            scoring='accuracy', 
                            cv=10,  
                            verbose=1) 
grid_search.fit(X_resampled, y_resampled)

print(f"Best hyperparameters: {grid_search.best_params_}")
print(grid_search.best_score_)
```

From above , I get the best model which has parameter like those 

Then , I check my tuning model is overfitting or not

```python
from sklearn.model_selection import learning_curve

rfc_best =  RandomForestClassifier(**grid_search.best_params_)

train_sizes, train_scores, test_scores = learning_curve(rfc_best,
                                                        X_resampled, y_resampled, cv=10, train_sizes=np.linspace(.1, 1.0, 5),scoring='accuracy')

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
plt.title('RFC')
plt.ylabel('Accuracy')
plt.xlabel('Sizes')
plt.legend(loc="best")
plt.show()
```

As you can see from graph above , there is no large gap between these two lines 

So I choose it as the best model for category prediction

```python
from sklearn.metrics import classification_report, confusion_matrix

from sklearn.model_selection import train_test_split

### Import and format data ###

scaler = MinMaxScaler()

smote = SMOTE(random_state=0)

X_scaled = scaler.fit_transform(X)

# Resample the dataset
X_resampled, y_resampled = smote.fit_resample(X_scaled, y.values.ravel())

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=0)

### Train the best model and try to test ### 

model_rfc = RandomForestClassifier(**grid_search.best_params_,
                                   random_state=0)

model_rfc.fit(X_train, y_train)

y_pred = model_rfc.predict(X_test)  # predicted probabilities of positive class

cm = confusion_matrix(y_test, y_pred)  # confusion matrix

plt.figure(figsize=(10, 8))
sns.heatmap(cm, 
            annot=True, 
            cmap='coolwarm',
            fmt='d',
            xticklabels=np.unique(y_test),
            yticklabels=np.unique(y_pred))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print(classification_report(y_test, y_pred))
print('-'*40)
```

```python
df = pd.read_csv('Customer Segmentation.csv')
df = df.drop('CustomerID',axis=1)
df = df[df['Category'].isnull()]
df.drop('Category',axis=1,inplace=True)
df
```

My features have some rows which is missing value , I handle with criteria below:

quantitative features : impute by median (because from above I used to test normality and I found each category in each quantitative feature is non-normal)

qualitative features : impute by most_frequent

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

# Separate quantitative and qualitative columns
quant_cols = ['Age','WorkExperience', 'FamilySize']
qual_cols = ['Gender', 'Married', 'Graduated', 'Profession','SpendingScore']

# Define transformers
quant_transformer = SimpleImputer(strategy='median')
qual_transformer = SimpleImputer(strategy='most_frequent')

# Apply transformers to columns
preprocessor = ColumnTransformer(transformers=[
    ('quant', quant_transformer, quant_cols),
    ('qual', qual_transformer, qual_cols)
])

# Impute missing values and transform data
X_preprocessor = preprocessor.fit_transform(df)

# Convert numpy array to pandas dataframe
df_imputed = pd.DataFrame(X_preprocessor, columns=quant_cols+qual_cols)
df_imputed
```

```python
df_imputed.dtypes
```

```python
df_imputed['Age'] = pd.to_numeric(df_imputed['Age'])
df_imputed['WorkExperience'] = pd.to_numeric(df_imputed['WorkExperience'])
df_imputed['FamilySize'] = pd.to_numeric(df_imputed['FamilySize'])
```

```python
df_imputed = pd.get_dummies(df_imputed)
df_imputed
```

```python
df_scaled = scaler.fit_transform(df_imputed)
df_scaled
```

```python
cat_pred = model_rfc.predict(df_scaled) 

df_pred = pd.read_csv('Customer Segmentation.csv')
df_pred = df_pred[df_pred['Category'].isnull()]
df_pred.drop('Category',axis=1,inplace=True)
df_pred.reset_index(inplace=True)
df_pred.drop('index',axis=1,inplace=True)
df_pred['Category'] = cat_pred
df_pred
```
